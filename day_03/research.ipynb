### 2. Comparison Operators (Create Boolean Values)

```python
x = 10
y = 5

# Equal to
print(x == y)     # False (10 is not equal to 5)

# Not equal to
print(x != y)     # True (10 is not equal to 5)

# Greater than
print(x > y)      # True (10 is greater than 5)

# Less than
print(x < y)      # False (10 is not less than 5)

# Greater than or equal to
print(x >= 10)    # True (10 is equal to 10)

# Less than or equal to
print(y <= 5)     # True (5 is equal to 5)
```

### 3. Logical Operators

### 1 .AND Operator (and)


Rule: Returns True only if BOTH conditions are True


```python
# Example 1
age = 25
has_license = True

can_drive = (age >= 18) and has_license
print(can_drive)  # True (both conditions are True)

# Example 2
age = 16
has_license = True

can_drive = (age >= 18) and has_license
print(can_drive)  # False (age condition is False)
```

###  B. OR Operator (or)

Rule: Returns True if AT LEAST ONE condition is True

```python
# Example 1
has_cash = False
has_card = True

can_pay = has_cash or has_card
print(can_pay)  # True (one condition is True)

# Example 2
has_cash = False
has_card = False

can_pay = has_cash or has_card
print(can_pay)  # False (both are False)
```
