#### 1. Root Node

The top node of the tree

Represents the first and most important decision

#### 2. Internal Nodes

Nodes that perform feature-based tests

### 3. Leaf Nodes

Terminal nodes

Store the final output

For:

Classification: Class label (e.g., Disease / No Disease)
